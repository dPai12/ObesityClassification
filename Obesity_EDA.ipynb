```python
df_orig = df_orig.rename(columns={'NObeyesdad':'target'})
```

```python
df_orig.isnull().values.sum()
```

```python
plt.figure(figsize=(10, 6))

# Using a Histogram
sns.histplot(df_orig['Age'], bins=30, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
t_order = ['Insufficient_Weight','Normal_Weight', 'Overweight_Level_I', 'Overweight_Level_II', 'Obesity_Type_I', 'Obesity_Type_II', 'Obesity_Type_III']
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Weight', hue='target', data=df_orig, hue_order = t_order)
plt.title('Scatter Plot of Age and Weight by Obesity Levels')
plt.show()
```

```python
df_orig['target'] = df_orig['target'].astype('category')
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(y='target', x='Age', data=df_orig, order = t_order)
plt.title('Average Age by Obesity Levels')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='target', x='Age', data=df_orig, order = t_order)
plt.title('Box Plot of Age by Obesity Level')
plt.show()

# plt.figure(figsize=(10, 6))
# sns.boxplot(y='target', x='Weight', data=df_orig, order = t_order)
# plt.title('Box Plot of Weight by Obesity Level')
# plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Weight', hue='Gender', data=df_orig)
plt.title('Scatter Plot of Age and Weight by Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(hue='Gender', x='target', data=df_orig, palette='viridis', edgecolor='black', dodge=False)
plt.title('Distribution of Obesity Level by Gender and Smoking Status')

plt.show()
```

```python
sns.countplot(x='Gender', data=df_orig, palette='viridis', edgecolor='black')

plt.title('Distribution of Gender')
plt.show()
```

```python
sns.violinplot(x='MTRANS', y='Weight', data=df_orig)
plt.title('Violin Plot of Weight by Transportation Mode')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='MTRANS', hue='target', data=df_orig, hue_order = t_order)
plt.title('Distribution of Target by Mode of Transportation')
plt.show()
```
